```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
mlr= pd.read_csv("C://Users//91948//Downloads//ass 5//50_Startups.csv")
mlr
```

distinguish between dependent and independent dataset

```python
X= mlr.iloc[:,:-1]
Y= mlr.iloc[:,4]
```

coverting categorical coloumn into one hot encoding

```python
states = pd.get_dummies(X['State'],drop_first=True)
```

dropping off the state coloumn

```python
X = X.drop('State',axis=1)
```

```python
X= pd.concat([X,states],axis= 1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor= LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
y_predict = regressor.predict(X_test)
```

```python
y_predict1 = regressor.predict(X_train)
```

```python
y_predict
```

```python
y_predict1
```

```python
from sklearn.metrics import r2_score
score= r2_score(Y_test,y_predict)
score
```

```python
final_mv= smf.ols("Price~age+KM+HP+cc+Doors+Gears+QT+Weight",data=cars_2).fit()
```

```python
(final_mv.rsquared,final_mv.aic)
```

```python
model_influence_V= final_mv.get_influence()
(c_v, _)=model_influence_V.cooks_distance
```

```python
fig= plt.subplots(figsize=(20,7))
plt.stem(np.arange(len(cars_2)),np.round(c_v,3));
plt.xlabel('Row INDEX')
plt.ylabel('cooks distance');
```
